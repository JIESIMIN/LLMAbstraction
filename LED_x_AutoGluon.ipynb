```python
import pandas as pd
from sklearn.model_selection import train_test_split
from google.colab import files

# Step 1: Load your uploaded file
df = pd.read_csv("LED_summaries_labeled.csv")

# Step 2: Split into 80% training and 20% testing
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)

# Step 3: Save to CSV
train_path = "LED_summaries_training.csv"
test_path = "LED_summaries_testing.csv"
train_df.to_csv(train_path, index=False)
test_df.to_csv(test_path, index=False)

# Step 4: Download the files
files.download(train_path)
files.download(test_path)
```

```python
import pandas as pd

# Get the filename from the uploaded dictionary
filename = list(training.keys())[0]

# Load the uploaded CSV file
train_data = pd.read_csv(filename)

# Preview
train_data.head()
```

```python
label = 'label'
train_data[label].describe()
```

```python
predictor = TabularPredictor(label=label).fit(train_data)
```

Prediction

```python
import pandas as pd

# Make sure 'testing' is a proper DataFrame
testing = pd.read_csv("LED_summaries_testing.csv")

# Define the label column
label = "label"  # Replace this with the actual column name

# Drop the label column to create the test features
X_test = testing.drop(columns=[label])

# Run prediction
y_pred = predictor.predict(X_test)

# Optional: View first few predictions
print(y_pred[:5])
```

Evaluation


```python
predictor.evaluate(testing, silent=True)
```
